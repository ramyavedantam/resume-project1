```python
cols=inp0.columns.to_list()
```

```python
cat_cols=['operating_system','storage_type','gpu_class','warranty','refurbished','brand']
```

```python
cont_cols=['price','RAM_nth','storage_nth','cpu_benchmark','screen_size','PPI']
```

```python
#####single variable analysis
```

```python
for col in cols:
    sns.histplot(inp0[col])
    plt.title(col)
    plt.show()
```

```python
for col in cont_cols:
    sns.boxplot(inp0[col])
    plt.title(col)
    plt.show()
```

```python
for col in cont_cols:
    sns.distplot(inp0[col])
    plt.title(col)
    plt.show()
```

```python
###bivariate analysis
```

```python
for col in cont_cols:
    for col2 in cat_cols:
        sns.scatterplot(x=inp0[col],y=inp0[col2])
        plt.title(price+" Vs "+col2)
        plt.show()
```

```python
 for col in cont_cols:
    for col2 in cont_cols:
        sns.scatterplot(x=inp0[col],y=inp0[col2])
        plt.title(price+" Vs "+col2)
        plt.show()
```
